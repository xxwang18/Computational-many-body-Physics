Exercise 1: TASEP (8 points) As shown in the lecture, the TASEP with parallel update corresponds to rule 184. Here we consider the TASEP with N = 50 sites and periodic boundary conditions.

a) Choose a random starting configuration with M particles (M ≈ 25) and calculate the average flow for Nt = 100 time steps. The flow is defined here as the number of particles per unit of time transferred from site N to site 1. (5 points)

```python
import numpy as np
n_cols = 50 #set columns of the configuration
n_particle = 25
n_gen = 100  # set number of generations
rule_n = 184 # set rule N
num_N_1 = 0 # the number of particle transfer from position N to 1

z = np.zeros((1,n_cols),dtype = int)
for i in range(n_particle):
    z[0][i] = 1
np.random.shuffle(z[0])#Creat the initial configuration
#
def rule (n):
    rule_list = [0,0,0,0,0,0,0,0]
    for i in range(8):
        rule_list[7-i] = n%2
        n=n//2
    return rule_list
rule(rule_n)
```

```python
# create the first n_gen generations of rule N

N_t = np.zeros(n_gen,dtype = int)
next_row = np.zeros(n_cols,dtype = int)
for row_i in range(n_gen):
    for i in range(n_cols):
        next_row [i]= rule(rule_n)[7-(z[row_i,(i-1+n_cols)%n_cols]*4+z[row_i,i]*2+z[row_i,(i+1+n_cols)%n_cols]*1)] 
        if next_row[i] == 1:
            N_t[row_i]=N_t[row_i]+1
    if z[row_i,-1]==1 and z[row_i,0]==0:
        num_N_1 +=1
    z = np.append(z,[next_row],axis=0)
num_N_1
```

```python
#show the picture of rule N
print(z)
import matplotlib.pyplot as plt
plt.title('TASEP: time dependence')
plt.xlabel('x')
plt.ylabel('t')
plt.imshow(z)
```

```python
average_flow = num_N_1/n_gen
average_flow
```
